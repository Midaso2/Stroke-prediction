# **🧹 3. Advanced Data Preprocessing Pipeline**

````python
# =============================================================================
# 🧹 ADVANCED DATA PREPROCESSING PIPELINE
# =============================================================================

print("🧹 ADVANCED DATA PREPROCESSING PIPELINE")
print("=" * 50)

# Track preprocessing steps
preprocessing_log = []

# Step 1: Remove unnecessary ID column if present
if 'id' in df.columns:
    df = df.drop('id', axis=1)
    preprocessing_log.append("✅ Removed 'id' column")
    print("✅ Removed 'id' column")

# Step 2: Handle 'Other' gender category (medical datasets often have this issue)
if 'gender' in df.columns:
    other_count = (df['gender'] == 'Other').sum() if 'Other' in df['gender'].values else 0
    if other_count > 0:
        df = df[df['gender'] != 'Other']
        preprocessing_log.append(f"⚠️ Removed {other_count} rows with 'Other' gender")
        print(f"⚠️ Removed {other_count} rows with 'Other' gender")
    else:
        preprocessing_log.append("✅ No 'Other' gender values found")
        print("✅ No 'Other' gender values found")

# Step 3: Enhanced categorical encoding with pandas warnings suppression
print("\n🔄 ENHANCED CATEGORICAL DATA ENCODING")
print("-" * 40)

# Suppress pandas FutureWarning for replace operations
pd.options.mode.chained_assignment = None

# Encode categorical variables with medical domain knowledge
encoding_mappings = {}

if 'gender' in df.columns:
    gender_mapping = {'Male': 0, 'Female': 1}
    df['gender'] = df['gender'].replace(gender_mapping).astype(int)
    encoding_mappings['gender'] = gender_mapping
    print("✅ Gender encoded: Male=0, Female=1")

if 'ever_married' in df.columns:
    marriage_mapping = {'No': 0, 'Yes': 1}
    df['ever_married'] = df['ever_married'].replace(marriage_mapping).astype(int)
    encoding_mappings['ever_married'] = marriage_mapping
    print("✅ Marriage status encoded: No=0, Yes=1")

if 'work_type' in df.columns:
    work_mapping = {'Private': 0, 'Self-employed': 1, 'Govt_job': 2, 'children': 3, 'Never_worked': 4}
    df['work_type'] = df['work_type'].replace(work_mapping).astype(int)
    encoding_mappings['work_type'] = work_mapping
    print("✅ Work type encoded:", work_mapping)

if 'Residence_type' in df.columns:
    residence_mapping = {'Urban': 0, 'Rural': 1}
    df['Residence_type'] = df['Residence_type'].replace(residence_mapping).astype(int)
    encoding_mappings['Residence_type'] = residence_mapping
    print("✅ Residence encoded: Urban=0, Rural=1")

if 'smoking_status' in df.columns:
    smoking_mapping = {'never smoked': 0, 'formerly smoked': 1, 'Unknown': 2, 'smokes': 3}
    df['smoking_status'] = df['smoking_status'].replace(smoking_mapping).astype(int)
    encoding_mappings['smoking_status'] = smoking_mapping
    print("✅ Smoking status encoded:", smoking_mapping)

# Store encoding mappings for future reference
preprocessing_log.append(f"✅ Categorical encoding completed with {len(encoding_mappings)} variables")

print(f"\n📊 ENCODING SUMMARY")
print("-" * 30)
print(f"Variables encoded: {len(encoding_mappings)}")
print(f"Total mappings created: {sum(len(mapping) for mapping in encoding_mappings.values())}")
````

# **🔬 4. Advanced Missing Value Treatment**

````python
# =============================================================================
# 🔬 ADVANCED MISSING VALUE TREATMENT
# =============================================================================

print("🔬 ADVANCED MISSING VALUE TREATMENT")
print("=" * 50)

# Analyze missing patterns before treatment
missing_before = df.isnull().sum()
print("Missing values before treatment:")
missing_data_before = missing_before[missing_before > 0]
if len(missing_data_before) > 0:
    for col, count in missing_data_before.items():
        print(f"  {col}: {count} ({count/len(df)*100:.1f}%)")
else:
    print("  ✅ No missing values detected")

# Advanced BMI missing value handling
if 'bmi' in df.columns:
    bmi_missing = df['bmi'].isnull().sum()
    if bmi_missing > 0:
        print(f"\n💊 BMI MISSING VALUE STRATEGY")
        print("-" * 30)
        
        # Analyze BMI distribution before imputation
        bmi_stats = df['bmi'].describe()
        print(f"BMI statistics (before imputation):")
        print(f"  Mean: {bmi_stats['mean']:.2f}")
        print(f"  Median: {bmi_stats['50%']:.2f}")
        print(f"  Std: {bmi_stats['std']:.2f}")
        
        # Use median imputation for more robust handling of outliers
        bmi_median = df['bmi'].median()
        df['bmi'] = df['bmi'].fillna(bmi_median)
        preprocessing_log.append(f"✅ BMI: {bmi_missing} missing values imputed with median ({bmi_median:.2f})")
        print(f"✅ Filled {bmi_missing} missing BMI values with median: {bmi_median:.2f}")
        print(f"   Rationale: Median chosen over mean to reduce outlier impact")
    else:
        print("✅ No missing BMI values detected")

# Check for any remaining missing values
remaining_missing = df.isnull().sum().sum()
print(f"\n🎯 MISSING VALUE TREATMENT COMPLETE")
print("-" * 40)
print(f"Remaining missing values: {remaining_missing}")

if remaining_missing == 0:
    print("✅ All missing values successfully treated")
    preprocessing_log.append("✅ All missing values successfully treated")
else:
    print(f"⚠️ {remaining_missing} missing values still require attention")
    remaining_by_col = df.isnull().sum()
    remaining_by_col = remaining_by_col[remaining_by_col > 0]
    for col, count in remaining_by_col.items():
        print(f"  {col}: {count} missing")
````

## **🔬 Machine Learning Pipeline Fundamentals**

### **📚 Core ML Pipeline Concepts**

**🎯 What is a Machine Learning Pipeline?**

A machine learning pipeline is a sequence of automated tasks that transform raw data into a trained model capable of making predictions. In healthcare applications like stroke prediction, this pipeline ensures:

- **Consistency**: Every data point follows the same preprocessing steps
- **Reproducibility**: Results can be replicated across different environments
- **Scalability**: New data can be processed automatically using the same workflow
- **Quality Assurance**: Each step validates and improves data quality

**🔄 Pipeline Components Overview:**

1. **Data Cleaning**: Handle missing values, outliers, and inconsistencies
2. **Feature Engineering**: Create new features from existing data
3. **Feature Scaling**: Normalize feature scales for algorithm compatibility
4. **Feature Selection**: Choose most relevant features for model training
5. **Model Training**: Apply machine learning algorithms to learn patterns
6. **Model Evaluation**: Assess performance using appropriate metrics

### **⚖️ Estimator Types in Scikit-learn**

**🔧 Transformer Estimators:**
- Use `.fit()` and `.transform()` methods
- Learn parameters from training data, then apply transformations
- Examples: `StandardScaler`, `SelectFromModel`, `PCA`
- Purpose: Data preprocessing and feature engineering

**🎯 Predictor Estimators:**

````python
# =============================================================================
# 🔬 ADVANCED MACHINE LEARNING PIPELINE DEVELOPMENT
# =============================================================================

print("🔬 ADVANCED MACHINE LEARNING PIPELINE DEVELOPMENT")
print("=" * 60)

# Advanced ML imports for comprehensive model development
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectFromModel
from sklearn.decomposition import PCA
from sklearn.metrics import make_scorer, recall_score, precision_score, f1_score
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Tree-based algorithms for comprehensive comparison
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier, ExtraTreesClassifier
try:
    from xgboost import XGBClassifier
    xgb_available = True
    print("✅ XGBoost available for advanced ensemble modeling")
except ImportError:
    xgb_available = False
    print("⚠️ XGBoost not available - install with: pip install xgboost")

# Prepare data for ML pipeline development
print(f"\n📊 PREPARING DATA FOR ADVANCED ML PIPELINE")
print("-" * 50)

# Verify target variable
if 'stroke' in df.columns:
    X = df.drop('stroke', axis=1)
    y = df['stroke']
    print(f"✅ Features: {X.shape[1]} variables")
    print(f"✅ Target: {y.value_counts().to_dict()}")
    print(f"✅ Class balance: {y.value_counts(normalize=True).round(3).to_dict()}")
else:
    print("❌ Target variable 'stroke' not found in dataset")
    raise ValueError("Cannot proceed without target variable")

# Professional train-test split with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2, 
    random_state=101, 
    stratify=y  # Maintain class balance in splits
)

print(f"\n📈 DATASET SPLIT SUMMARY")
print("-" * 30)
print(f"Training set: {X_train.shape[0]:,} samples ({X_train.shape[0]/len(df)*100:.1f}%)")
print(f"Test set: {X_test.shape[0]:,} samples ({X_test.shape[0]/len(df)*100:.1f}%)")
print(f"Train stroke rate: {y_train.mean()*100:.1f}%")
print(f"Test stroke rate: {y_test.mean()*100:.1f}%")
print("✅ Stratified split maintains class balance")
````

````python
# =============================================================================
# 🏗️ ADVANCED PIPELINE FACTORY FOR MULTIPLE ALGORITHMS
# =============================================================================

print("\n🏗️ ADVANCED PIPELINE FACTORY DEVELOPMENT")
print("=" * 50)

def create_pipeline(algorithm_name, algorithm, include_pca=False, n_components=None):
    """
    Creates advanced ML pipeline with configurable preprocessing steps.
    
    Parameters:
    -----------
    algorithm_name : str
        Name of the algorithm for logging
    algorithm : sklearn estimator
        The machine learning algorithm
    include_pca : bool, default=False
        Whether to include PCA dimensionality reduction
    n_components : int or None, default=None
        Number of PCA components (if None, uses 95% variance)
    
    Returns:
    --------
    pipeline : sklearn.Pipeline
        Configured pipeline ready for training
    """
    
    steps = []
    
    # Step 1: Feature Scaling (essential for many algorithms)
    steps.append(('scaler', StandardScaler()))
    
    # Step 2: Dimensionality Reduction (optional)
    if include_pca:
        if n_components is None:
            # Automatically determine components for 95% variance
            pca_step = PCA(n_components=0.95, random_state=101)
        else:
            pca_step = PCA(n_components=n_components, random_state=101)
        steps.append(('pca', pca_step))
        print(f"  ✅ {algorithm_name}: PCA dimensionality reduction included")
    
    # Step 3: Feature Selection (tree-based algorithms can handle this)
    if hasattr(algorithm, 'feature_importances_'):
        feature_selector = SelectFromModel(
            algorithm.__class__(random_state=101), 
            threshold='median'
        )
        steps.append(('feature_selection', feature_selector))
        print(f"  ✅ {algorithm_name}: Feature selection included")
    
    # Step 4: Final Algorithm
    steps.append(('model', algorithm))
    
    pipeline = Pipeline(steps)
    print(f"  🔧 {algorithm_name}: Pipeline created with {len(steps)} steps")
    
    return pipeline

# Define algorithm configurations for comprehensive comparison
algorithm_configs = {
    'Decision Tree': {
        'algorithm': DecisionTreeClassifier(random_state=101),
        'include_pca': False,
        'description': 'Single tree with interpretable decision rules'
    },
    'Random Forest': {
        'algorithm': RandomForestClassifier(random_state=101),
        'include_pca': False,
        'description': 'Ensemble of trees with bootstrap aggregating'
    },
    'Gradient Boosting': {
        'algorithm': GradientBoostingClassifier(random_state=101),
        'include_pca': False,
        'description': 'Sequential tree building with gradient optimization'
    },
    'AdaBoost': {
        'algorithm': AdaBoostClassifier(random_state=101),
        'include_pca': False,
        'description': 'Adaptive boosting with weighted sample training'
    },
    'Extra Trees': {
        'algorithm': ExtraTreesClassifier(random_state=101),
        'include_pca': False,
        'description': 'Extremely randomized trees with random splits'
    }
}

# Add XGBoost if available
if xgb_available:
    algorithm_configs['XGBoost'] = {
        'algorithm': XGBClassifier(random_state=101, eval_metric='logloss'),
        'include_pca': False,
        'description': 'Extreme gradient boosting with regularization'
    }

print(f"\n📚 ALGORITHM CONFIGURATION SUMMARY")
print("-" * 40)
for name, config in algorithm_configs.items():
    print(f"{name:20s}: {config['description']}")

print(f"\n✅ Pipeline factory ready for {len(algorithm_configs)} algorithms")
````

````python
# =============================================================================
# 🏥 COMPREHENSIVE CLINICAL PERFORMANCE EVALUATION
# =============================================================================

print("\n🏥 COMPREHENSIVE CLINICAL PERFORMANCE EVALUATION")
print("=" * 60)

def clinical_confusion_matrix_analysis(y_true, y_pred, algorithm_name):
    """
    Provides clinical interpretation of confusion matrix for stroke prediction.
    """
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    
    print(f"\n🔍 {algorithm_name} - Clinical Confusion Matrix Analysis")
    print("-" * 50)
    
    # Create DataFrame for better visualization
    cm_df = pd.DataFrame(cm, 
                        columns=['Predicted: No Stroke', 'Predicted: Stroke'],
                        index=['Actual: No Stroke', 'Actual: Stroke'])
    display(cm_df)
    
    # Clinical interpretation
    print(f"\nClinical Interpretation:")
    print(f"  ✅ True Negatives (Correct No-Stroke): {tn:,}")
    print(f"  ⚠️ False Positives (False Alarms): {fp:,}")
    print(f"  🚨 False Negatives (Missed Strokes): {fn:,}")
    print(f"  ✅ True Positives (Detected Strokes): {tp:,}")
    
    # Clinical implications
    if fn > 0:
        print(f"  ⚠️ CRITICAL: {fn} stroke cases missed - requires clinical review")
    else:
        print(f"  ✅ EXCELLENT: No stroke cases missed")
    
    if fp > 10:
        print(f"  📊 MODERATE: {fp} false alarms - may increase unnecessary testing")
    else:
        print(f"  ✅ GOOD: Low false alarm rate")
    
    return cm_df

def comprehensive_classification_report(y_true, y_pred, algorithm_name):
    """
    Generates comprehensive classification report with clinical focus.
    """
    print(f"\n📊 {algorithm_name} - Comprehensive Performance Metrics")
    print("-" * 50)
    
    # Standard classification report
    report = classification_report(y_true, y_pred, 
                                 target_names=['No Stroke', 'Stroke'],
                                 output_dict=True)
    
    # Key metrics for clinical decision making
    no_stroke_precision = report['No Stroke']['precision']
    no_stroke_recall = report['No Stroke']['recall']
    stroke_precision = report['Stroke']['precision']
    stroke_recall = report['Stroke']['recall']
    overall_accuracy = report['accuracy']
    
    print(f"Overall Model Performance:")
    print(f"  Accuracy: {overall_accuracy:.3f}")
    print(f"  Macro Average F1: {report['macro avg']['f1-score']:.3f}")
    print(f"  Weighted Average F1: {report['weighted avg']['f1-score']:.3f}")
    
    print(f"\nStroke Detection Performance (Class 1):")
    print(f"  Precision: {stroke_precision:.3f} (of predicted strokes, {stroke_precision*100:.1f}% are correct)")
    print(f"  Recall: {stroke_recall:.3f} (of actual strokes, {stroke_recall*100:.1f}% are detected)")
    print(f"  F1-Score: {report['Stroke']['f1-score']:.3f}")
    
    print(f"\nNo-Stroke Classification Performance (Class 0):")
    print(f"  Precision: {no_stroke_precision:.3f}")
    print(f"  Recall: {no_stroke_recall:.3f}")
    print(f"  F1-Score: {report['No Stroke']['f1-score']:.3f}")
    
    # Clinical recommendations
    print(f"\nClinical Recommendations:")
    if stroke_recall >= 0.90:
        print("  ✅ EXCELLENT stroke detection - suitable for clinical screening")
    elif stroke_recall >= 0.80:
        print("  ✅ GOOD stroke detection - acceptable for clinical use with monitoring")
    elif stroke_recall >= 0.70:
        print("  ⚠️ MODERATE stroke detection - requires improvement for clinical deployment")
    else:
        print("  🚨 POOR stroke detection - not suitable for clinical use")
    
    if stroke_precision >= 0.50:
        print("  ✅ Acceptable precision - reasonable false alarm rate")
    else:
        print("  ⚠️ Low precision - high false alarm rate may burden healthcare system")
    
    return report

def feature_importance_analysis(pipeline, feature_names, algorithm_name):
    """
    Analyzes and visualizes feature importance for tree-based models.
    """
    if not hasattr(pipeline['model'], 'feature_importances_'):
        print(f"  ⚠️ {algorithm_name}: Feature importance not available for this algorithm")
        return None
    
    print(f"\n🌟 {algorithm_name} - Feature Importance Analysis")
    print("-" * 50)
    
    # Get feature names after preprocessing
    if 'feature_selection' in pipeline.named_steps:
        selected_features = feature_names[pipeline['feature_selection'].get_support()]
    else:
        selected_features = feature_names
    
    # Get feature importances
    importances = pipeline['model'].feature_importances_
    
    # Create importance DataFrame
    importance_df = pd.DataFrame({
        'Feature': selected_features,
        'Importance': importances
    }).sort_values('Importance', ascending=False)
    
    # Display top features
    print("Top 5 Most Important Features:")
    for i, (idx, row) in enumerate(importance_df.head().iterrows(), 1):
        print(f"  {i}. {row['Feature']}: {row['Importance']:.4f}")
    
    # Plot feature importance
    plt.figure(figsize=(10, 6))
    plt.barh(range(len(importance_df)), importance_df['Importance'])
    plt.yticks(range(len(importance_df)), importance_df['Feature'])
    plt.xlabel('Feature Importance')
    plt.title(f'{algorithm_name} - Feature Importance')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
    
    return importance_df

print("✅ Clinical evaluation functions ready")
print("📊 Proceeding with comprehensive model evaluation...")
````

````python
# =============================================================================
# 🔍 PCA DIMENSIONALITY REDUCTION ANALYSIS
# =============================================================================

print("\n🔍 PCA DIMENSIONALITY REDUCTION ANALYSIS")
print("=" * 60)

def analyze_pca_variance_explained(X_scaled, max_components=None):
    """
    Analyzes PCA variance explained to determine optimal number of components.
    """
    if max_components is None:
        max_components = min(X_scaled.shape[1], 15)  # Limit for visualization
    
    print(f"📊 Analyzing PCA variance for up to {max_components} components")
    
    # Fit PCA with all components
    pca_full = PCA(n_components=max_components, random_state=101)
    pca_full.fit(X_scaled)
    
    # Calculate cumulative variance explained
    variance_explained = pca_full.explained_variance_ratio_
    cumulative_variance = np.cumsum(variance_explained)
    
    # Create DataFrame for analysis
    pca_analysis = pd.DataFrame({
        'Component': [f'PC{i+1}' for i in range(max_components)],
        'Individual_Variance': variance_explained,
        'Cumulative_Variance': cumulative_variance,
        'Variance_Percentage': variance_explained * 100,
        'Cumulative_Percentage': cumulative_variance * 100
    })
    
    print(f"\nPCA Variance Analysis (Top 10 Components):")
    display(pca_analysis.head(10))
    
    # Find components for different variance thresholds
    components_80 = np.where(cumulative_variance >= 0.80)[0][0] + 1
    components_90 = np.where(cumulative_variance >= 0.90)[0][0] + 1
    components_95 = np.where(cumulative_variance >= 0.95)[0][0] + 1
    
    print(f"\nComponents needed for variance thresholds:")
    print(f"  80% variance: {components_80} components")
    print(f"  90% variance: {components_90} components")
    print(f"  95% variance: {components_95} components")
    
    # Plot variance explained
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Individual variance
    ax1.bar(range(1, max_components + 1), variance_explained * 100)
    ax1.set_xlabel('Principal Component')
    ax1.set_ylabel('Variance Explained (%)')
    ax1.set_title('Individual Variance Explained by Each Component')
    ax1.grid(True, alpha=0.3)
    
    # Cumulative variance
    ax2.plot(range(1, max_components + 1), cumulative_variance * 100, 'b-o')
    ax2.axhline(y=80, color='r', linestyle='--', label='80% threshold')
    ax2.axhline(y=90, color='g', linestyle='--', label='90% threshold')
    ax2.axhline(y=95, color='orange', linestyle='--', label='95% threshold')
    ax2.set_xlabel('Number of Components')
    ax2.set_ylabel('Cumulative Variance Explained (%)')
    ax2.set_title('Cumulative Variance Explained')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return pca_analysis, components_80, components_90, components_95

def visualize_pca_stroke_separation(X_scaled, y, n_components=3):
    """
    Visualizes how PCA components separate stroke vs non-stroke cases.
    """
    print(f"\n🎨 PCA STROKE PATTERN VISUALIZATION")
    print("-" * 40)
    
    # Apply PCA
    pca = PCA(n_components=n_components, random_state=101)
    X_pca = pca.fit_transform(X_scaled)
    
    # Calculate variance explained by selected components
    variance_explained = pca.explained_variance_ratio_
    total_variance = variance_explained.sum()
    
    print(f"Variance explained by {n_components} components: {total_variance*100:.1f}%")
    for i, var in enumerate(variance_explained):
        print(f"  PC{i+1}: {var*100:.1f}%")
    
    # 2D visualization
    plt.figure(figsize=(15, 5))
    
    # Plot 1: PC1 vs PC2
    plt.subplot(1, 3, 1)
    colors = ['blue', 'red']
    labels = ['No Stroke', 'Stroke']
    for i, (color, label) in enumerate(zip(colors, labels)):
        mask = y == i
        plt.scatter(X_pca[mask, 0], X_pca[mask, 1], 
                   c=color, alpha=0.6, label=label, s=30)
    plt.xlabel(f'PC1 ({variance_explained[0]*100:.1f}% variance)')
    plt.ylabel(f'PC2 ({variance_explained[1]*100:.1f}% variance)')
    plt.title('Stroke Cases in PC1-PC2 Space')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Plot 2: PC1 vs PC3
    if n_components >= 3:
        plt.subplot(1, 3, 2)
        for i, (color, label) in enumerate(zip(colors, labels)):
            mask = y == i
            plt.scatter(X_pca[mask, 0], X_pca[mask, 2], 
                       c=color, alpha=0.6, label=label, s=30)
        plt.xlabel(f'PC1 ({variance_explained[0]*100:.1f}% variance)')
        plt.ylabel(f'PC3 ({variance_explained[2]*100:.1f}% variance)')
        plt.title('Stroke Cases in PC1-PC3 Space')
        plt.legend()
        plt.grid(True, alpha=0.3)
    
    # Plot 3: PC2 vs PC3
    if n_components >= 3:
        plt.subplot(1, 3, 3)
        for i, (color, label) in enumerate(zip(colors, labels)):
            mask = y == i
            plt.scatter(X_pca[mask, 1], X_pca[mask, 2], 
                       c=color, alpha=0.6, label=label, s=30)
        plt.xlabel(f'PC2 ({variance_explained[1]*100:.1f}% variance)')
        plt.ylabel(f'PC3 ({variance_explained[2]*100:.1f}% variance)')
        plt.title('Stroke Cases in PC2-PC3 Space')
        plt.legend()
        plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Analyze separation quality
    stroke_mask = y == 1
    no_stroke_mask = y == 0
    
    print(f"\nPCA Separation Analysis:")
    print(f"  Stroke cases: {stroke_mask.sum():,}")
    print(f"  No-stroke cases: {no_stroke_mask.sum():,}")
    
    # Calculate mean positions for each class
    stroke_mean = X_pca[stroke_mask].mean(axis=0)
    no_stroke_mean = X_pca[no_stroke_mask].mean(axis=0)
    
    # Calculate distance between class centers
    center_distance = np.linalg.norm(stroke_mean - no_stroke_mean)
    print(f"  Distance between class centers: {center_distance:.3f}")
    
    return X_pca, pca

# Prepare scaled data for PCA analysis
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("✅ Data scaled and ready for PCA analysis")
````
