# **🚀 1. Enhanced Data Acquisition with Kaggle Hub Integration**

```python
# =============================================================================
# 📊 ADVANCED DATA ACQUISITION WITH KAGGLE HUB INTEGRATION
# =============================================================================

# Install and import Kaggle Hub for automated dataset download
try:
    import kagglehub
    print("✅ Kaggle Hub already installed")
except ImportError:
    print("⚠️ Installing Kaggle Hub...")
    import subprocess
    subprocess.check_call(['pip', 'install', 'kagglehub'])
    import kagglehub
    print("✅ Kaggle Hub installation completed")

# Download latest stroke prediction dataset
print("📥 Downloading stroke prediction dataset from Kaggle...")
try:
    path = kagglehub.dataset_download("fedesoriano/stroke-prediction-dataset")
    print(f"📂 Dataset path: {path}")
    use_kaggle_data = True
except Exception as e:
    print(f"⚠️ Kaggle download failed: {e}")
    print("📁 Will use local dataset")
    use_kaggle_data = False

# Enhanced imports for advanced analysis
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set professional plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("📚 Enhanced libraries imported successfully")
```

```python
# =============================================================================
# 📁 ROBUST DATA LOADING WITH ERROR HANDLING
# =============================================================================

# Load dataset with comprehensive error handling
if use_kaggle_data:
    try:
        file_name = 'healthcare-dataset-stroke-data.csv'
        file_path = os.path.join(path, file_name)
        
        if not os.path.exists(file_path):
            # Try alternative filename
            files = os.listdir(path)
            csv_files = [f for f in files if f.endswith('.csv')]
            if csv_files:
                file_name = csv_files[0]
                file_path = os.path.join(path, file_name)
            else:
                raise FileNotFoundError("No CSV files found in dataset directory")
        
        df = pd.read_csv(file_path)
        print(f"✅ Kaggle dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")
        data_source = "Kaggle Hub"
        
    except Exception as e:
        print(f"⚠️ Error loading Kaggle dataset: {e}")
        use_kaggle_data = False

if not use_kaggle_data:
    print("📁 Loading local dataset...")
    df = pd.read_csv('../Stroke.csv')
    print(f"✅ Local dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")
    data_source = "Local File"

# Store original dataset for comparison
df_original = df.copy()

print(f"\n📊 DATA SOURCE: {data_source}")
print(f"📏 Dataset dimensions: {df.shape}")
print(f"💾 Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

# **🔍 2. Comprehensive Data Quality Assessment**

```python
# =============================================================================
# 🔍 COMPREHENSIVE DATA QUALITY ASSESSMENT
# =============================================================================

print("🔍 COMPREHENSIVE DATA QUALITY ASSESSMENT")
print("=" * 50)

# 1. Basic dataset information
print("📋 DATASET OVERVIEW")
print("-" * 30)
print(f"Rows: {df.shape[0]:,}")
print(f"Columns: {df.shape[1]}")
print(f"Data types: {df.dtypes.value_counts().to_dict()}")

# 2. Display first few rows
print("\n📊 SAMPLE DATA (First 5 rows):")
print("-" * 30)
display(df.head())

# 3. Missing value analysis
print("\n🔍 MISSING VALUE ANALYSIS")
print("-" * 30)
missing_counts = df.isnull().sum()
missing_percentages = (missing_counts / len(df)) * 100
missing_summary = pd.DataFrame({
    'Missing_Count': missing_counts,
    'Missing_Percentage': missing_percentages
}).round(2)

# Show only columns with missing values
missing_data = missing_summary[missing_summary['Missing_Count'] > 0]
if len(missing_data) > 0:
    print("Columns with missing values:")
    display(missing_data)
else:
    print("✅ No missing values detected in initial assessment")

# 4. Data type validation
print("\n🔢 DATA TYPE VALIDATION")
print("-" * 30)
print("Column data types:")
for col in df.columns:
    print(f"{col}: {df[col].dtype}")

# 5. Numerical data validation
print("\n🔢 NUMERICAL DATA VALIDATION")
print("-" * 30)
numerical_cols = df.select_dtypes(include=[np.number]).columns
if len(numerical_cols) > 0:
    inf_counts = np.isinf(df[numerical_cols]).sum()
    print("Infinite values per numerical column:")
    if inf_counts.sum() > 0:
        print(inf_counts[inf_counts > 0])
    else:
        print("✅ No infinite values detected")
else:
    print("⚠️ No numerical columns detected")

# 6. Categorical data exploration
print("\n📊 CATEGORICAL DATA EXPLORATION")
print("-" * 30)
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    unique_vals = df[col].unique()
    print(f"{col}: {unique_vals[:10]}{'...' if len(unique_vals) > 10 else ''}")
    print(f"  → Unique values: {len(unique_vals)}")

# 7. Target variable analysis
print(f"\n🎯 TARGET VARIABLE ANALYSIS")
print("-" * 30)
if 'stroke' in df.columns:
    stroke_dist = df['stroke'].value_counts()
    stroke_pct = df['stroke'].value_counts(normalize=True) * 100
    print(f"No Stroke (0): {stroke_dist[0]:,} ({stroke_pct[0]:.1f}%)")
    print(f"Stroke (1): {stroke_dist[1]:,} ({stroke_pct[1]:.1f}%)")
    print(f"Class imbalance ratio: {stroke_dist[0]/stroke_dist[1]:.1f}:1")
    print(f"⚠️ Class imbalance detected - will require specialized handling")
else:
    print("⚠️ Target variable 'stroke' not found")
```

# **🧹 3. Advanced Data Preprocessing Pipeline**

```python
# =============================================================================
# 🧹 ADVANCED DATA PREPROCESSING PIPELINE
# =============================================================================

print("🧹 ADVANCED DATA PREPROCESSING PIPELINE")
print("=" * 50)

# Track preprocessing steps
preprocessing_log = []

# Step 1: Remove unnecessary ID column if present
if 'id' in df.columns:
    df = df.drop('id', axis=1)
    preprocessing_log.append("✅ Removed 'id' column")
    print("✅ Removed 'id' column")

# Step 2: Handle 'Other' gender category (medical datasets often have this issue)
if 'gender' in df.columns:
    other_count = (df['gender'] == 'Other').sum() if 'Other' in df['gender'].values else 0
    if other_count > 0:
        df = df[df['gender'] != 'Other']
        preprocessing_log.append(f"⚠️ Removed {other_count} rows with 'Other' gender")
        print(f"⚠️ Removed {other_count} rows with 'Other' gender")
    else:
        preprocessing_log.append("✅ No 'Other' gender values found")
        print("✅ No 'Other' gender values found")

# Step 3: Enhanced categorical encoding with pandas warnings suppression
print("\n🔄 ENHANCED CATEGORICAL DATA ENCODING")
print("-" * 40)

# Suppress pandas FutureWarning for replace operations
pd.options.mode.chained_assignment = None

# Encode categorical variables with medical domain knowledge
encoding_mappings = {}

if 'gender' in df.columns:
    gender_mapping = {'Male': 0, 'Female': 1}
    df['gender'] = df['gender'].replace(gender_mapping).astype(int)
    encoding_mappings['gender'] = gender_mapping
    print("✅ Gender encoded: Male=0, Female=1")

if 'ever_married' in df.columns:
    marriage_mapping = {'No': 0, 'Yes': 1}
    df['ever_married'] = df['ever_married'].replace(marriage_mapping).astype(int)
    encoding_mappings['ever_married'] = marriage_mapping
    print("✅ Marriage status encoded: No=0, Yes=1")

if 'work_type' in df.columns:
    work_mapping = {'Private': 0, 'Self-employed': 1, 'Govt_job': 2, 'children': 3, 'Never_worked': 4}
    df['work_type'] = df['work_type'].replace(work_mapping).astype(int)
    encoding_mappings['work_type'] = work_mapping
    print("✅ Work type encoded:", work_mapping)

if 'Residence_type' in df.columns:
    residence_mapping = {'Urban': 0, 'Rural': 1}
    df['Residence_type'] = df['Residence_type'].replace(residence_mapping).astype(int)
    encoding_mappings['Residence_type'] = residence_mapping
    print("✅ Residence encoded: Urban=0, Rural=1")

if 'smoking_status' in df.columns:
    smoking_mapping = {'never smoked': 0, 'formerly smoked': 1, 'Unknown': 2, 'smokes': 3}
    df['smoking_status'] = df['smoking_status'].replace(smoking_mapping).astype(int)
    encoding_mappings['smoking_status'] = smoking_mapping
    print("✅ Smoking status encoded:", smoking_mapping)

# Store encoding mappings for future reference
preprocessing_log.append(f"✅ Categorical encoding completed with {len(encoding_mappings)} variables")

print(f"\n📊 ENCODING SUMMARY")
print("-" * 30)
print(f"Variables encoded: {len(encoding_mappings)}")
print(f"Total mappings created: {sum(len(mapping) for mapping in encoding_mappings.values())}")
```

# **🔬 4. Advanced Missing Value Treatment**

```python
# =============================================================================
# 🔬 ADVANCED MISSING VALUE TREATMENT
# =============================================================================

print("🔬 ADVANCED MISSING VALUE TREATMENT")
print("=" * 50)

# Analyze missing patterns before treatment
missing_before = df.isnull().sum()
print("Missing values before treatment:")
missing_data_before = missing_before[missing_before > 0]
if len(missing_data_before) > 0:
    for col, count in missing_data_before.items():
        print(f"  {col}: {count} ({count/len(df)*100:.1f}%)")
else:
    print("  ✅ No missing values detected")

# Advanced BMI missing value handling
if 'bmi' in df.columns:
    bmi_missing = df['bmi'].isnull().sum()
    if bmi_missing > 0:
        print(f"\n💊 BMI MISSING VALUE STRATEGY")
        print("-" * 30)
        
        # Analyze BMI distribution before imputation
        bmi_stats = df['bmi'].describe()
        print(f"BMI statistics (before imputation):")
        print(f"  Mean: {bmi_stats['mean']:.2f}")
        print(f"  Median: {bmi_stats['50%']:.2f}")
        print(f"  Std: {bmi_stats['std']:.2f}")
        
        # Use median imputation for more robust handling of outliers
        bmi_median = df['bmi'].median()
        df['bmi'] = df['bmi'].fillna(bmi_median)
        preprocessing_log.append(f"✅ BMI: {bmi_missing} missing values imputed with median ({bmi_median:.2f})")
        print(f"✅ Filled {bmi_missing} missing BMI values with median: {bmi_median:.2f}")
        print(f"   Rationale: Median chosen over mean to reduce outlier impact")
    else:
        print("✅ No missing BMI values detected")

# Check for any remaining missing values
remaining_missing = df.isnull().sum().sum()
print(f"\n🎯 MISSING VALUE TREATMENT COMPLETE")
print("-" * 40)
print(f"Remaining missing values: {remaining_missing}")

if remaining_missing == 0:
    print("✅ All missing values successfully treated")
    preprocessing_log.append("✅ All missing values successfully treated")
else:
    print(f"⚠️ {remaining_missing} missing values still require attention")
    remaining_by_col = df.isnull().sum()
    remaining_by_col = remaining_by_col[remaining_by_col > 0]
    for col, count in remaining_by_col.items():
        print(f"  {col}: {count} missing")
```

# **📊 5. Data Validation and Quality Verification**

```python
# =============================================================================
# 📊 DATA VALIDATION AND QUALITY VERIFICATION
# =============================================================================

print("📊 DATA VALIDATION AND QUALITY VERIFICATION")
print("=" * 50)

# 1. Verify data types after preprocessing
print("🔍 DATA TYPE VERIFICATION")
print("-" * 30)
print("Final data types:")
type_summary = df.dtypes.value_counts()
for dtype, count in type_summary.items():
    print(f"  {dtype}: {count} columns")

# 2. Check for data consistency
print("\n✅ DATA CONSISTENCY CHECKS")
print("-" * 30)

consistency_issues = []

# Check age ranges
if 'age' in df.columns:
    age_min, age_max = df['age'].min(), df['age'].max()
    print(f"Age range: {age_min:.1f} - {age_max:.1f} years")
    if age_min < 0 or age_max > 120:
        consistency_issues.append(f"⚠️ Unusual age values: {age_min}-{age_max}")
    else:
        print("  ✅ Age values within expected range")

# Check BMI ranges
if 'bmi' in df.columns:
    bmi_min, bmi_max = df['bmi'].min(), df['bmi'].max()
    print(f"BMI range: {bmi_min:.1f} - {bmi_max:.1f}")
    if bmi_min < 10 or bmi_max > 60:
        consistency_issues.append(f"⚠️ Extreme BMI values: {bmi_min}-{bmi_max}")
    else:
        print("  ✅ BMI values within reasonable range")

# Check glucose levels
if 'avg_glucose_level' in df.columns:
    glucose_min, glucose_max = df['avg_glucose_level'].min(), df['avg_glucose_level'].max()
    print(f"Glucose range: {glucose_min:.1f} - {glucose_max:.1f} mg/dL")
    if glucose_min < 50 or glucose_max > 400:
        consistency_issues.append(f"⚠️ Extreme glucose values: {glucose_min}-{glucose_max}")
    else:
        print("  ✅ Glucose levels within clinical range")

# 3. Display consistency issues if any
if consistency_issues:
    print("\n⚠️ DATA CONSISTENCY ISSUES:")
    for issue in consistency_issues:
        print(f"  {issue}")
else:
    print("\n✅ No data consistency issues detected")

# 4. Final dataset summary
print(f"\n📊 FINAL PREPROCESSED DATASET SUMMARY")
print("-" * 40)
print(f"Final dimensions: {df.shape}")
print(f"Rows removed during preprocessing: {df_original.shape[0] - df.shape[0]}")
print(f"Columns processed: {df.shape[1]}")
print(f"Missing values: {df.isnull().sum().sum()}")
print(f"Duplicate rows: {df.duplicated().sum()}")

# 5. Display sample of processed data
print(f"\n🔍 PROCESSED DATA SAMPLE (First 10 rows)")
print("-" * 40)
display(df.head(10))
```
